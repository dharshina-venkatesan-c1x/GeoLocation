```python
#Extracting UserIP and UUID data from click_log
import pandas as pd
from ip2geotools.databases.noncommercial import DbIpCity

click_log = pd.read_excel("/Users/dharshina/Downloads/C1X/172.31.4.99_jsonclick.log copy.xlsx")
data = click_log[['UUID','UserIP']]
data
```

```python
#Converting the given IP to City and appending the data into a List
ip_city = []
for index,row in data.iterrows():
    response = DbIpCity.get(row[1])
    num_city = response.city
    ip_city.append(num_city)
    df_city = pd.DataFrame(ip_city)
df_city
```

```python
#Converting IP to Country and inserting the data into a List
ip_country = []
for index,row in data.iterrows():
    response = DbIpCity.get(row[1])
    num_country = response.country
    ip_country.append(num_country)
    df_country = pd.DataFrame(ip_country)
df_country
```

```python
#GeoLocation: Creating a Excel file with UUID, UserIp, Country, City
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile

df = pd.concat([data, df_country, df_city], axis = 1)

# changing columns using .columns() 
df.columns = ['UUID', 'UserIP', 'Country', 'City'] 
  
writer = ExcelWriter('GeoLocationData.xlsx')
df.to_excel(writer,'Sheet1',index=False)
writer.save()
```

```python
#Count of UUID in a city
pd.value_counts(df.City)
```
